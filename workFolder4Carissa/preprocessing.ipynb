```python
import requests
from bs4 import BeautifulSoup
import os
from urllib.parse import urlparse
```

```python
def fetch_privacy_policy(url):
    try:
        html = requests.get(url)
        html.raise_for_status()  # Raises an HTTPError for bad responses
        
        soup = BeautifulSoup(html.content, 'html.parser')

        last_marked = soup.find_all('div', {'class': 'p-law__inner'})[-1]
        result = last_marked.get_text()
        result = result.replace('\n', ' ')
        result = result.replace('\r', ' ')
        result = result.replace('\t', ' ')     

        return result
    except Exception as e:
        return f"An error occurred: {e}"
```

```python
def convert_to_file(string, website_name)
    domain = urlparse(website_name).netloc
    file = open(f'{domain}.html', "w")
    file.write(string)
    return file
```
